##Tokenising with spaCy

```python
import en_core_web_sm
nlp = en_core_web_sm.load()
```

```python
document = nlp(joenlp)
```

```python
# @title
document.text[0:500]
```

```python
tokens = [word.text for word in document]
tokens
```

##Write code to get the first 20 tokens

```python
tokens[0:20]
```

```python
sentences = [sent.text for sent in document.sents]
sentences
```

Normalising to Lowercases

```python
tokens_lower = [token.lower() for token in tokens]
tokens_lower[0:20]
```

##**Removing** Punctuation

```python
# Import a modle that helps with string processing
import string

# Make a table that translates all punctuation to an empty valu('None')
table = str.maketrans('','', string.punctuation)
punc_table = {chr(key):value for (key, value) in table.items()}
punc_table
```

```python
tokens_nopunct = [token.translate(table) for token in tokens_lower]
tokens_nopunct[0:20]
```

**Removing Non-Word Tokens**

```python
tokens_notempty = [token for token in tokens_nopunct if token != '']
tokens_notempty[0:10]
```

```python
# we can remove all the newline characters by adding a condition that filters out all non-alphabetic characters.
# The string method to use is isalpha()

tokens = [token for token in tokens_notempty if token.isalpha()]
tokens
```
